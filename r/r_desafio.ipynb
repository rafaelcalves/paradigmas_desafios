### Exemplo 1

```r
d = read.table('lotr_data.csv', sep=',', header=T, fill=T, quote="\"")
d
```

```r
plot(d$gender)
```

```r
unique(d$gender) #valores distintos
table(d$gender) #frequencia dos valores
```

```r
# normaliza os gêneros
d$gender <- as.character(d$gender)
d$gender[d$gender %in% c("Males", "male", "Most likely male")]<- "Male"
d$gender[d$gender == ""] <- "Undefined"
d$gender <- as.factor(d$gender)

unique(d$gender) #valores distintos
table(d$gender) #frequencia dos valores
plot(d$gender)
```

### Exemplo 2

```r
unique(d$race) 
table(d$race)
```

```r
# normalizar as raças que se referem a homens
d$race <- as.character(d$race)
d$race[d$race %in% unique(d$race[grepl("Men", d$race)])] <- "Men"
d$race <- as.factor(d$race)

unique(d$race) 
table(d$race)
```
